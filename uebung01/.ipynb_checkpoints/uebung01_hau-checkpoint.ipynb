````python
import spacy
import en_core_web_sm
import os
import matplotlib
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup

nlp = en_core_web_sm.load()
````

````python
def all_filenames(path):
    """ recursivly read all filenames into list """
    if not os.path.exists(path):
        print("path does not exist")
        return False
    
    dirs = []
    files = []
    for entry in os.listdir(path):
        subdir_path = os.path.join(path, entry)
        is_hidden = entry.startswith('.')
        is_xml = entry.endswith('.xml')     
        if os.path.isdir(subdir_path):
            dirs.append(subdir_path) 
        if os.path.isfile(subdir_path) and not is_hidden and is_xml:
            files.append(subdir_path)
            
    if len(dirs) > 0:
        for subdir_path in dirs:
            sub_files = all_filenames(subdir_path)
            files += sub_files
    
    return files


def read_all_xml_files(path):
    files = []
    for filename in all_filenames(path):
        content = read_file(filename)
        files.append((filename, content))
    return files

    
def read_file(path):
    with open(path, 'r') as f:
        try:
            return f.read()
        except UnicodeDecodeError:
            print(path)
            
            
def save_files(files, save_path):
    for (dir_path, content) in files:
        with open(save_path, 'w') as f:
            f.write(content)
    
                
def tokenize_add_PoS(path):
    files = read_all_xml_files(path)
    res = []
    for (filename, content) in files:
        soup = BeautifulSoup(content, 'xml')
        doc = nlp(soup.get_text().replace("\n", ""))
        current = []
        for token in doc:
            current.append((token.text, token.pos_))
        res.append(current)
            
    return res
````

````python
tokenize_add_PoS('./Training/')
````

````python
def count_PoS_tags(filenames):
    res = {}
    for filename in filenames:
        file = read_file(filename)
        soup = BeautifulSoup(file, 'xml')
        doc = nlp(soup.get_text().replace("\n", " "))
        
        for token in doc:
            pos = token.pos_
            if pos in res:
                res[pos] += 1
            else:
                res[pos] = 1
                
    return res
````

````python
filenames = all_filenames('./Training/')
count_PoS_tags(filenames)
````

````python
def count_entities(filenames):
    res = {
        'PLACE': 0,
        'SPATIAL_ENTITY': 0,
        'MOTION': 0,
        'MOTION_SIGNAL': 0,
        'SPATIAL_SIGNAL': 0,
        'QSLINK': 0,
        'OLINK': 0,
    }
    for filename in filenames:
        file = read_file(filename)
        soup = BeautifulSoup(file, 'xml')
        for key in res.keys():
            res[key] += len(soup.find_all(key))
    return res
````

````python
count_entities(filenames)
````

````python
def count_QsLinks(filenames):
    res = {}
    for filename in filenames:
        file = read_file(filename)
        soup = BeautifulSoup(file, 'xml')
        for qs_link in soup.find_all('QSLINK'):
            attrs = qs_link.attrs
            relType = attrs['relType']
            if relType in res:
                res[relType] += 1
            else: 
                res[relType] = 0
    return res
````

````python
count_QsLinks(filenames)
````

````python
def count_length(filenames):
    res = {}
    for filename in filenames:
        file = read_file(filename)
        soup = BeautifulSoup(file, 'xml')
        without_tags = soup.get_text()
        for sentence in without_tags.split('.'):
            length = len(sentence)
            if length == 0: 
                continue
            if length in res:
                res[length] += 1
            else:
                res[length] = 1
    return res


def plot_length_and_num(length_and_num):
    x = length_and_num.keys()
    y = length_and_num.values()
    fig, ax = plt.subplots()
    plt.bar(x, y)
    ax.set(xlabel="Satzlänge", ylabel="Häufigkeit")
````

````python
length_and_num = count_length(filenames)
plot_length_and_num(length_and_num)
````

````python
def get_triggers(filenames):
    """
    qs_links = 
        { id of SPATIAL_SIGNAL : [ids of QSLINK that triggered] }
    o_links  =
        { id of SPATIAL_SIGNAL : [ids of OLINK that triggered] }
    """
    res = {
        'qs_links': {},
        'o_links': {}
    }
    for filename in filenames:
        file = read_file(filename)
        soup = BeautifulSoup(file, 'xml')
        
        qs_links = res['qs_links']
        for link in soup.find_all('QSLINK'):
            attrs = link.attrs
            trigger = attrs['trigger']
            if trigger in qs_links:
                qs_links[trigger].append(attrs['id'])
            else:
                qs_links[trigger] = [attrs['id']]
                
        o_links = res['o_links']
        for link in soup.find_all('OLINK'):
            attrs = link.attrs
            trigger = attrs['trigger']
            if trigger in o_links:
                o_links[trigger].append(attrs['id'])
            else:
                o_links[trigger] = [attrs['id']]
        
    return res      
````

````python
get_triggers(filenames)
````
